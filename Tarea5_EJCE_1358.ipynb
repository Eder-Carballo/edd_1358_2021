```
class NodoDoble:
    def __init__( self , value , previo = None , siguiente = None ):
        self.data = value
        self.prev = previo
        self.next = siguiente

class DoubleLinkedList:
    def __init__( self ):
        self.__head = None
        self.__tail = None
        self.__size = 0 

    def get_size( self ):
        return self.__size

    def is_empty( self ):
        return self.__size ==0

    def append( self , value ):
        if self.is_empty():
            nuevo = NodoDoble( value )
            self.__head = nuevo
            self.__tail = nuevo 
        else: 
            nuevo = NodoDoble( value , self.__tail , None )
            self.__tail.next = nuevo
            self.__tail = nuevo

        self.__size += 1

    def find_from_tail( self , value ):
        pos = 0 
        curr_node = self.__tail
        while curr_node.data != value:
            curr_node = curr_node.prev
            pos += 1
        print(f"El valor {value} buscado a partir de tail se encuentra en la posicion: {pos}")

    def find_from_head( self , value ):
        pos = 0 
        curr_node = self.__head
        while curr_node.data != value:
            curr_node = curr_node.next
            pos += 1
        print(f"El valor {value} buscado a partir de head se encuentra en la posicion: {pos}")
        
    def remove_from_tail( self , value ):
        if self.is_empty():
            pass
        elif self.__tail.prev == None:
            self.__tail = self.__head = None
            self.__size = 0
        else:
            curr_node = self.__tail
            while curr_node.data != value and curr_node != None:
                curr_node = curr_node.prev
            if curr_node.data == value:
                curr_node.next.prev = curr_node.prev
                curr_node.prev.next = curr_node.next
        self.__size -= 1 

    def remove_from_head( self , value ):
        anterior = None
        if self.is_empty():
            pass
        elif self.__head.next == None:
            self.__head = self.__tail = None
            self.__size = 0
        else:
            curr_node = self.__head
            while curr_node.data != value and curr_node != None:
                curr_node = curr_node.next
            if curr_node.data == value and curr_node.data == self.__head.data:
                anterior = curr_node
                self.__head = self.__head.next
            elif curr_node.data == value:
                curr_node.prev.next = curr_node.next
                curr_node.next.prev = curr_node.prev
        self.__size -= 1 

    def transversal( self ):
        curr_node = self.__head
        while curr_node != None:
            print("<---" , curr_node.data , "--->" , end="" )
            curr_node = curr_node.next
        print("")

    def reverse_transversal( self ):
        curr_node = self.__tail
        while curr_node != None:
            print("<---" , curr_node.data , "--->" , end="" )
            curr_node = curr_node.prev
        print("")
```

```
ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)
ldl.append(60)
ldl.append(70)
ldl.append(80)
ldl.append(90)
ldl.append(100)

ldl.transversal()
print("")

ldl.reverse_transversal()
print("")

ldl.remove_from_head(10)
ldl.transversal()
print("")

ldl.remove_from_tail(40)
ldl.transversal()
print("")

ldl.find_from_head(70)
print("")

ldl.find_from_tail(30)
```
