```
class Nodo:

    def __init__(self , dato , siguiente=None ):
        self.dato = dato
        self.siguiente = siguiente

class CircularList:
    def __init__( self ):
        self.__ref = None
        self.__size = 0

    def is_empty( self ):
        return self.__ref == None

    def insert( self , value ):
        curr_node = self.__ref
        nuevo = Nodo( value )
        if self.__ref == None:
            self.__ref = nuevo
            nuevo.siguiente = self.__ref
        else:
            if nuevo.dato > curr_node.dato:
                while curr_node.siguiente != self.__ref:
                    curr_node = curr_node.siguiente
                curr_node.siguiente = nuevo
                nuevo.siguiente = self.__ref
                self.__ref = nuevo
            else:
                if nuevo.dato < curr_node.siguiente.dato:
                    while curr_node.siguiente != self.__ref:
                        curr_node = curr_node.siguiente             
                    curr_node.siguiente = nuevo
                    nuevo.siguiente = self.__ref
                    #self.__ref = nuevo
                
    def travel(self):
                 #Recorriendo la lista vinculada
        if self.__ref == None:
            return
        curr_node = self.__ref
        print(curr_node.dato)
        while curr_node.siguiente != self.__ref:
            curr_node = curr_node.siguiente
            print(curr_node.dato)

    def search( self , value ):
        curr_node = self.__ref 
        while curr_node.siguiente != self.__ref and value != curr_node.dato:
            curr_node = curr_node.siguiente
        return print(value == curr_node.dato)

    def remove( self , value ):
        if self.__ref == None:
            return
        elif value == self.__ref.dato:
            curr_node = self.__ref
            while curr_node.siguiente != self.__ref:
                curr_node = curr_node.siguiente
            curr_node.siguiente = self.__ref.siguiente
            self.__ref = self.__ref.siguiente
        else:
            curr_node = self.__ref
            pre = None
            while curr_node.dato != value:
                pre = curr_node
                curr_node = curr_node.siguiente
            pre.siguiente = curr_node.siguiente
```
